# Use clustering algorithm(K-means) to divide the total data by the Top 10 most important features into several similar groups

```python
CLUSTER_NUMS = 11
TEST_SIZE = 0.3
RANDOM_STATE = 42

# Top 10 most important features
TOP_IMPORTANCE = ['density',
 'space_group',
 'volume',
 'MagpieData range Electronegativity',
 'formation_energy_per_atom',
 'density_atomic',
 'MagpieData range Number',
 'MagpieData maximum Number',
 'MagpieData minimum Electronegativity',
 'MagpieData range CovalentRadius'][0:2]
```

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
```

```python
# read data
data = pd.read_csv('../../data/domain_based_data.csv')
```

```python
data.head()
```

```python
X = data.drop(['e_electronic', 'e_ionic', 'e_total'], axis=1)
y = data[['e_electronic', 'e_ionic', 'e_total']]
display(X.head())
display(y.head())
```

- Train and Test set split

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)
```

```python
kmeans = KMeans(n_clusters=CLUSTER_NUMS, random_state=RANDOM_STATE)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_cluster = X_train[TOP_IMPORTANCE]
X_cluster_scaled = scaler.fit_transform(X_cluster)
```

```python
kmeans.fit(X_cluster_scaled)
```

```python
X_train['cluster'] = kmeans.labels_
```

```python
cluster_test = X_test[TOP_IMPORTANCE]
cluster_test_scaled = scaler.transform(cluster_test)
```

```python
# predict clusters for test data
X_test['cluster'] = kmeans.predict(cluster_test_scaled)
```

```python
X_test.head()
```

```python
groups = X_train.groupby('cluster')
```

```python
for cluster, df in groups:
    print(f'Cluster {cluster}: {len(df)} samples')
    # display(df.head(5))
```

- Visualize the clusters

```python
X_temp = X_train.copy()
X_temp['e_electronic'] = y_train['e_electronic']
X_temp['e_total'] = data.iloc[X_temp.index, :]['e_total']
# visualize clusters
# use seaborn to visualize clusters
import seaborn as sns
plt.figure(figsize=(10, 6), dpi=400)
sns.scatterplot(data=X_temp, x='density', y='space_group', hue='cluster', palette=sns.color_palette('muted',n_colors=CLUSTER_NUMS))
# plt.xscale('log')
# plt.yscale('log')
# cluster based on the two most important features
plt.title('clusters based on the two most important features')
plt.xlabel('density($g/cm^3$)')
plt.ylabel('space_group')

# # save picture
# plt.savefig('../../figures/domain_based/clusters_ionic.png')
# plt.show()
```

```python
# Merge X_train and y_train
train_set = X_train.copy()
train_set[['e_electronic', 'e_ionic', 'e_total']] = y_train
# Merge X_test and y_test
test_set = X_test.copy()
test_set[['e_electronic', 'e_ionic', 'e_total']] = y_test
```

```python
train_set['class'] = 'train'
test_set['class'] = 'test'
set = pd.concat([train_set, test_set])
set.head()
```
