# 数据分析1
- 分开训练e_electronic和e_ionic, 再将其相加起来进行预测整个的e_total(方法一)，
与直接训练e_total(方法二)的准确度比较

- **结果：** 并没有很大的差别

```python
# 导库
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# 导入数据
e_total_pred = pd.read_csv('../data/groups/test_value/e_total_pred.csv')
e_total_pred.head()
```

```python
# 导入数据
e_ionic_pred = pd.read_csv('../data/groups/test_value/e_ionic_pred.csv')
e_ionic_pred.head()
```

```python
# 导入数据
e_electronic_pred = pd.read_csv('../data/groups/test_value/e_electronic_pred.csv')
e_electronic_pred.head()
```

```python
# 合并数据
e_electronic_pred['e_ionic_pred'] = e_ionic_pred['e_ionic_pred']
e_electronic_pred['e_total_pred'] = e_total_pred['e_total_pred']
```

```python
e_pred = e_electronic_pred
e_pred.head()
```

```python
e_pred['eletronic_pred+ionic_pred'] = e_pred['e_ionic_pred'] + e_pred['e_electronic_pred']
e_pred.head()
```

```python
def plot_cross_val_scores(mse, mae, y_hat, y_train):
    '''绘制交叉验证的结果'''
    print(f'MSE: {mse}, MAE: {mae}')

    # 绘制parity plot
    fig, ax = plt.subplots(dpi=300)
    ax.scatter(y_train, y_hat, s=1)
    ax.plot(y_train, y_train, c='k', zorder=-5)
    ax.set_xlabel('Actual $\epsilon_{electronic}$')
    ax.set_ylabel('Predicted $\epsilon_{electronic}$')
    # ax.set_xscale('log')
    # ax.set_yscale('log')
    ax.set_xlim(0, 50)
    ax.set_ylim(0, 50)
    plt.show()
```

```python
# 方法一

# 计算MSE
mse_1 = mean_squared_error(e_pred['e_total'], e_pred['e_total_pred'])
# 计算MAE
mae_1 = mean_absolute_error(e_pred['e_total'], e_pred['e_total_pred'])

plot_cross_val_scores(mse_1, mae_1, e_pred['e_total_pred'], e_pred['e_total'])
```

```python
# 方法二

# 计算MSE
mse_2 = mean_squared_error(e_pred['e_total'], e_pred['eletronic_pred+ionic_pred'])
# 计算MAE
mae_2 = mean_absolute_error(e_pred['e_total'], e_pred['eletronic_pred+ionic_pred'])

plot_cross_val_scores(mse_2, mae_2, e_pred['eletronic_pred+ionic_pred'], e_pred['e_total'])
```
